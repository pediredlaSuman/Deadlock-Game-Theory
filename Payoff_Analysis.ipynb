```python
import matplotlib.pyplot as plt
import seaborn as sns

# Results data
strategy_names = ["Tit for Tat", "Sample Strategy", "Friedman Strategy", "Joss Sneaky", "Graaskamp", "Tester"]
computer_payoffs = [3, 3, 3, 3, 3, 0]
player_payoffs = [3, 3, 3, 3, 3, 3]

# Create a bar chart for computer payoffs
plt.figure(figsize=(10, 6))
sns.barplot(x=strategy_names, y=computer_payoffs, palette="viridis")
plt.title("Computer Payoffs by Strategy")
plt.xlabel("Strategy")
plt.ylabel("Computer Payoff")
plt.xticks(rotation=45)
plt.show()

# Create a bar chart for player payoffs
plt.figure(figsize=(10, 6))
sns.barplot(x=strategy_names, y=player_payoffs, palette="magma")
plt.title("Player Payoffs by Strategy")
plt.xlabel("Strategy")
plt.ylabel("Player Payoff")
plt.xticks(rotation=45)
plt.show()
```
