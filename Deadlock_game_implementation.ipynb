```python
import random

# Function fir playing the Deadlock Game.
def play_deadlock_game():
    print("Welcome to the Deadlock Game!")
    while True:
        print("\nChoose your action:")
        print("1. Cooperate (C)")
        print("2. Defect (D)")
        print("3. Exit")
        user_choice = input("Your choice: ").strip().lower()

        # Exiting the game if the user chooses to exit.
        if user_choice == '3' or user_choice.lower() == 'exit':
            print("Exiting the game.")
            break

        # Handling invalid choices.
        if user_choice not in ['1', '2', 'c', 'd']:
            print("Invalid choice. Please choose again.")
            continue

        # Determining user action based on choice.
        user_action = 'C' if user_choice in ['1', 'c'] else 'D'
        computer_action = random.choice(['C', 'D'])

        # Determining payoffs based on actions.
        if user_action == 'C' and computer_action == 'C':
            user_payoff = 1
            computer_payoff = 1
        elif user_action == 'D' and computer_action == 'D':
            user_payoff = 2
            computer_payoff = 2
        else:
            if user_action == 'C':
                user_payoff = 0
                computer_payoff = 3
            else:
                user_payoff = 3
                computer_payoff = 0

        # Displaying actions and payoffs.
        print(f"\nYour action: {user_action}")
        print(f"Computer's action: {computer_action}")
        print(f"Your payoff: {user_payoff}")
        print(f"Computer's payoff: {computer_payoff}")

# Calling the main function to start the game.
if __name__ == "__main__":
    play_deadlock_game()
```
