```python
import matplotlib.pyplot as plt
import numpy as np
#excuting principles of game theory concpets in deadlock game
# Rational Decision-Making
players = ['Player A', 'Player B']
payoffs = {'Cooperate': [1, 1], 'Defect': [3, 0]}
colors = {'Cooperate': 'blue', 'Defect': 'red'}

plt.figure(figsize=(10, 6))
for action, payoff in payoffs.items():
    plt.bar(players, payoff, color=colors[action], label=action)
plt.xlabel('Players')
plt.ylabel('Payoffs')
plt.title('Payoff Matrix for Rational Decision-Making')
plt.legend()
plt.show()

# Strategic Interactions
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y1, label='Player A')
plt.plot(x, y2, label='Player B')
plt.xlabel('Time')
plt.ylabel('Strategy')
plt.title('Strategic Interactions Over Time')
plt.legend()
plt.show()

# Nash Equilibrium
nash_eq = (3, 4)  # Example Nash equilibrium
plt.figure(figsize=(8, 8))
plt.scatter(*nash_eq, color='green', label='Nash Equilibrium')
plt.xlabel('Player A Strategy')
plt.ylabel('Player B Strategy')
plt.title('Nash Equilibrium')
plt.legend()
plt.show()

# Cooperative and non-cooperative strategies
strategies = ['Cooperate', 'Non-Cooperate']
percentages = [60, 40]  # Example percentages of cooperative and non-cooperative strategies

plt.figure(figsize=(8, 6))
plt.pie(percentages, labels=strategies, autopct='%1.1f%%', colors=['blue', 'red'])
plt.title('Distribution of Cooperative and Non-Cooperative Strategies')
plt.show()

# Bargaining and Negotiation
bargaining_power = {'Player A': 0.7, 'Player B': 0.3}  # Example bargaining power distribution
plt.figure(figsize=(8, 6))
plt.bar(bargaining_power.keys(), bargaining_power.values(), color=['blue', 'red'])
plt.xlabel('Players')
plt.ylabel('Bargaining Power')
plt.title('Bargaining Power Distribution')
plt.show()
```
